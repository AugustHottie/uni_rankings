```python
import pandas as pd
import numpy as np
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
import openpyxl
import os
```

```python
df = pd.read_csv('./Book1.csv') 
```

```python
pd.set_option('display.max_columns', None)
```

### Here, I slice out the years 2022 and 2023

```python
rows_cut = df.drop(df[(df.Year == 2022) | (df.Year == 2023)].index)
```

```python
df = rows_cut
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

Here, we replaced all NaN values in the `rank` column with the median.

```python
df['rank'].replace(12345, np.nan, inplace=True)
df['rank'].fillna(df['rank'].quantile(0.50), inplace=True)
```

### Changed the data type of rank from float64 to int64

```python
df['rank'] = df['rank'].astype('int64')
```

### Here, I replace all NaN values with the 25th quantile

```python
df['scores_overall'].replace(12345, np.nan, inplace=True)
df['scores_overall'].replace(0.0, np.nan, inplace=True)
df['scores_overall'].fillna(df['scores_overall'].quantile(0.25), inplace=True)
```

### Changed the data type of the `scores_industry_income` column from an object to numeric
NaN values were then replaced with the mean, and rounded accordingly

```python
df['scores_industry_income'] = pd.to_numeric(df['scores_industry_income'], errors='coerce')
df['scores_industry_income'] .replace('-', df['scores_industry_income'].mean(), inplace=True)
```

```python
# df['scores_industry_income'].replace(0.0, np.nan, inplace=True)
df['scores_industry_income'].fillna(df['scores_industry_income'].mean(), inplace=True)
df['scores_industry_income'] = df['scores_industry_income'].round(2)
```

### The rows with wrong data type needed a fix, and NaN replaced with accordingly.
This was done after careful analysis of the data

```python
df['scores_teaching'].replace(np.nan, df['scores_teaching'].quantile(0.25), inplace=True)
```

```python
df['scores_international_outlook'] = pd.to_numeric(df['scores_international_outlook'], errors='coerce')
```

```python
df['scores_international_outlook'].replace('-', df['scores_international_outlook'].mean(), inplace=True)
```

```python
df['scores_international_outlook'] = df['scores_international_outlook'].astype('float64')
```

```python
df['scores_international_outlook'].fillna(0, inplace=True)
```
